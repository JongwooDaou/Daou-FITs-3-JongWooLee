```python
# 필요한 라이브러리 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#랜덤넘버 고정
np.random.seed(42)
```

```python
#고객수
n = 1000
```

```python
#고객 데이터 생성
data = {

    'customer_id': np.arange(1, n+1),
    'age': np.random.randint(18, 70, size = n),
    'tenure': np.random.randint(1, 10, n),
    'balance': np.random.randint(0, 25000, n),
    'num_products': np.random.randint(1, 3, n),
    'has_credit_card': np.random.choice([0, 1], n),
    'is_active_member': np.random.choice([0, 1], n),
    'estimated_salary': np.random.randint(30000, 150000, n),
    'churn': np.random.choice([0, 1], n, p=[0.8, 0.2])
}
#data frame 생성
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('customer_churn.csv', index=False)
df.info()
```

```python
# 결측지 확인
df.isnull().sum()
```

```python
# x and y 분리
y = df['churn']
x = df.drop('churn', axis=1)
```

```python
# 범주형 변수 인코딩
x = pd.get_dummies(x, drop_first=True)
```
